```python
#SCROLL_PAUSE_TIME = 0.2
for i in company_name:
    url='https://finance.yahoo.com/'
    driver.get(url)
    search=driver.find_element(By.CSS_SELECTOR,"input[name='yfin-usr-qry']")
    search.send_keys(i)
    search.send_keys(Keys.RETURN)
    sleep(1)
    driver.find_element(By.XPATH,'//*[@id="quote-nav"]/ul/li[5]').click()
    sleep(1)
    driver.find_element(By.XPATH,'//*[@id="Col1-1-HistoricalDataTable-Proxy"]/section/div[1]/div[1]/div[1]/div/div/div').click()
    inp=driver.find_element(By.CSS_SELECTOR,"input[name='startDate']")
    inp.send_keys('05-11-2018')
    inp.send_keys(Keys.RETURN)
    driver.find_element(By.XPATH,'//*[@id="dropdown-menu"]/div/div[3]/button[1]').click()
    driver.find_element(By.XPATH,'//*[@id="Col1-1-HistoricalDataTable-Proxy"]/section/div[1]/div[1]/button').click()
    
    from selenium.common.exceptions import NoSuchElementException
    date=[]
    opn=[]
    high=[]
    low=[]
    close=[]
    for j in range(1,1001):
        try:
            driver.execute_script("window.scrollTo(0,70000);")
            d=driver.find_element(By.XPATH,('//*[@id="Col1-1-HistoricalDataTable-Proxy"]/section/div[2]/table/tbody/tr['+str(j)+']/td[1]/span'))
            date.append(d.text)
            o=driver.find_element(By.XPATH,('//*[@id="Col1-1-HistoricalDataTable-Proxy"]/section/div[2]/table/tbody/tr['+str(j)+']/td[2]/span'))
            opn.append(o.text)
            h=driver.find_element(By.XPATH,('//*[@id="Col1-1-HistoricalDataTable-Proxy"]/section/div[2]/table/tbody/tr['+str(j)+']/td[3]/span'))
            high.append(h.text)
            l=driver.find_element(By.XPATH,('//*[@id="Col1-1-HistoricalDataTable-Proxy"]/section/div[2]/table/tbody/tr['+str(j)+']/td[4]/span'))
            low.append(l.text)
            c=driver.find_element(By.XPATH,('//*[@id="Col1-1-HistoricalDataTable-Proxy"]/section/div[2]/table/tbody/tr['+str(j)+']/td[5]/span'))
            close.append(c.text)
            #time.sleep(SCROLL_PAUSE_TIME)
            print(d.text,o.text,h.text,l.text,c.text)
        except NoSuchElementException:
            date.append(None)
            opn.append(None)
            high.append(None)
            low.append(None)
            close.append(None)        
        sleep(.05)
    df_temp = pd.DataFrame(list(zip(date,opn,high,low,close)),columns =['Date', 'Open','High','Low','Close'])
    df_temp['Company']=i
    df=pd.concat([df,df_temp],axis=0)
```
